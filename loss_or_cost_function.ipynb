<h3 style='color:blue'>Implement Mean Absolute Error</h3>

```python
def mae(y_predicted, y_true):
    total_error = 0
    for yp, yt in zip(y_predicted, y_true):
        total_error += abs(yp - yt)
    print("Total error is:",total_error)
    mae = total_error/len(y_predicted)
    print("Mean absolute error is:",mae)
    return mae
```

```python
float(mae(y_predicted, y_true))
```

```python
def log_loss(y_true, y_predicted):
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

<h3 style='color:blue'>Implement Mean Squared error</h3>

```python
import numpy as np

y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
def mse(y_true, y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted):
        total_error += (yt-yp)**2
    print("Total Squared Error:",total_error)
    mse = total_error/len(y_true)
    print("Mean Squared Error:",mse)
    return mse
```

```python
mse(y_true, y_predicted)
```

<h3 style='color:blue'>Implement Categorical Cross Entropy</h3>

```python
import numpy as np
import math

y_true = np.array([
    [1,0,0],
    [0,1,0],
    [0,0,1]
])

y_predicted = np.array([
    [0.8,0.1,0.1],
    [0.2,0.7,0.1],
    [0.1,0.2,0.7]
])

def categorical_cross_entropy(y_true, y_predicted):
    total_error = 0
    for yt, yp in zip(y_true, y_predicted):
        for i in range(len(yt)):
            total_error += -(yt[i] * math.log(yp[i]))
    print("Total Error:", total_error)
    cce = total_error / len(y_true)
    print("Categorical Cross Entropy:", cce)
    return cce

categorical_cross_entropy(y_true, y_predicted)
```

<h3 style='color:blue'>Implement Sparse Categorical Cross Entropy</h3>

```python
import numpy as np
import math

y_true = np.array([0,1,2])

y_predicted = np.array([
    [0.8,0.1,0.1],
    [0.2,0.7,0.1],
    [0.1,0.2,0.7]
])

def sparse_categorical_cross_entropy(y_true, y_predicted):
    total_error = 0
    for i in range(len(y_true)):
        total_error += -math.log(y_predicted[i][y_true[i]])
    print("Total Error:", total_error)
    scce = total_error / len(y_true)
    print("Sparse Categorical Cross Entropy:", scce)
    return scce

sparse_categorical_cross_entropy(y_true, y_predicted)
```
